## Chapter 1

# Introduction

발명자들은 생각할 수 있는 기계를 만들어내고자 하는 꿈을 오랬동안 꿔왔다. 이러한 욕망은 적어도 고대 그리스로 부터 시작되었다. 신화적인 존재인, 피그말리온, 다이달로스, 헤파이스토스는 전설적인 발명자로서 설명될 수 있을 것이고, 갈라테아, 탈로스, 판도라는 인공적인 생명체로 볼 수 있을 것이다 (Ovid and Martin,2004; Sparkes, 1996; Tandy, 1997). 

프로그래밍이 가능한 컴퓨터가 처음으로 등장 했을때, 사람들은 이러한 기계들이 지능적이 될 수 있을지에 대하여 100년여간, 첫번째 그러한 기계(Lovelace, 1842)가 탄생하기 전까지, 의문을 가졌다. 오늘날, 인공지능(AI)는 번성하는 분야 중 하나로, 수많은 실제 문제를 다루는 어플리케이션과 활발한 연구 주제에 적용되고 있다. 우리는 지능적인 소프트웨어가 일상적인 노동을 자동화 하는 것, 이미지나 음성을 이해하는것, 의학적 진단을 내리고 기본적인 과학적 연구를 할 수 있는 것으로 생각한다.

초창기 인공지능 분야는 사람들이 지능적으로 해결하기에 어려운 문제이지만, 상대적으로 형식적이고 수학적인 규정(공식)의 목록인 문제,를 빠르게 접근하고 해결하는데 사용되었다. 인공지능 분야에서의 진정한 도전 과제는 사람들이 쉽게 해결할 수 있는 작업이지만, 사람들이 일반화하여 설명하기 힘든 문제를 해결하는 것이다. 이러한 말하는 단어를 인식하거나 이미지에서 얼굴을 인지하는 것과 같은 문제를, 사람들은 직관적으로, 느낌상 자동적으로 해결한다.

이 책은 이러한 직관적인 문제들을 해결하기 위한 것을 다룬다. 제시하는 해결책은 컴퓨터가 경험을 통해서 배우고, 간단한 개념들의 관계로 부터 계층적 개념으로 세상을 이해하게 해준다. 경험으로 부터 지식을 획득함으로써, 이러한 접근 방법은 사람이 컴퓨터에게 필요한 모든 지식을 일일이 지정해주는 작업을 피할 수 있게 된다. 계층적 개념들은 컴퓨터가 보다 간단한 개념으로부터 좀더 복잡한 개념을 학습할 수 있는 능력을 부여한다. 이러한 개념들이 각각에 어떻게 계층적으로 연결되어 있는지 그래프를 그려번다면, 그 그래프는 아주 깊고 많은 계층을 가질 것이다. 이러한 이유로, 우리는 이러한 접근 방법을 AI deep learning (인공지능 딥러닝) 이라고 부른다.

과거의 AI 분야에서의 많은 성공들은 상대적으로 무익하고, 형식적인 환경에서 이루어 졌고, 컴퓨터가 세상에 대하여 많은 지식을 가지는 것에대해서 크게 요구하지 않았다. 예를 들어보면, IBM의 Deep Blue라는 체스를 두는 시스템은 1997년에 세계 챔피언인 Garry Kasparov를 상대로 승리를 거두었다. 체스는 단순히 64개의 움직일 수 있는 위치와 경직스럽게 국한적으로 움직일 수 있는 32개의 말로 이루어진 간단한 세상을 표현한다. 성공적인 체스게임의 전력을 고안하는 것은 엄청난 성취이지만, 이러한 성취에 대한 도전은 체스를 표현하고 컴퓨터에게 체스말을 움직이게 하는 것에 대한 어려움 때문이 아니다. 체스는 아주 간략하면서 완전한 규칙의 목록으로 설명될 수 있으며, 프로그래머에 의해서 몇수 앞선 시간에 대하여 이 목록의 일부가 제공될 수 있다.

아이러니 하게도, 사람들에게 정신적으로 힘든, 추상적이면서 규정화된 작업들은 컴퓨터에게는 가장 쉬운 일이다. 컴퓨터는 오랫동안 최고의 사람 체스 플레이어를 상대로 이겨 왔지만, 최근에 와서야 겨우 일반인들이 해낼 수 있는 사물과 음성을 인식하는 능력을 습득하기 시작했다. 한 사람의 일상은 세계에 대한 엄청나게 많은 지식을 요구한다. 이러한 지식의 대부분은 주관적이고 직관적이기 때문에, 어떤 규정화된 방법으로 이를 표현하기가 어렵다. 컴퓨터는 지능적인 방법으로 생동하기 위해서 이런 동일한 지능을 습득할 필요가 있다. 인공지능에서의, 가장 중요한 도전과제 중 하나는 어떻게 이러한 규정이 없는 지식을 컴퓨터에게 전달 할 수 있느냐 이다.

몇몇 인공지능 프로젝트들은 세상에 대한 지식을 규정화된 언어로서 하드-코드하여 찾곤 하였다. 컴퓨터는 논리적인 추론방법을 사용하여 이러한 규정화된 언어에서의 문장을 자동화하여 추리할 수 있다. 이 방법은 인공지능에 대한 기식 기반의 접근방법으로 알려져 있다. 이러한 프로젝트의 어떤 것도 중요한 성공으로 연결되지는 못했다. 참고로, 가장 유명한 프로젝트는 Cyc (Lenat and Guham 1989) 이다. Cyc는 추론 엔진이자 CycL라고 불리는 언어에 대한 문장의 데이터베이스이고, 이러한 문장들은 사람 감독자로 부터 입력 되었다. 꽤나 다루기 어려운 프로세스여서, 사람들은 세계를 정확하게 표현하기에 충분히 복잡한 규정화된 규칙을 생각해내는데 어려움을 겪었다. 예를 들어서, Cyc는 Fred라는 사람이 아침에 면도를 하는 이야기를 이해하는데 실패 했다. 여기에 사용된 추론 엔진은 이야기에서 모순을 감지했다: 사람들이 전자기기의 부분으로 이루어지지 않았음을 알았지만, Fred가 면도기를 들고 있었기 때문에, 이 시스템은 이 개체를 전자기기적인 부분이 포함된 "면도중인Fred" 라고 믿었다. 그래서, 면도중에도 Fred가 여전히 사람인지에 대해서 질문을 하였다.
